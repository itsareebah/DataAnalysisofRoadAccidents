```python
import pandas as pd
import numpy as np
```

```python
timeSeries = 'time of accident.xls'
df2 = pd.read_excel(timeSeries, index_col=None)
m = df2.columns
n = df2[df2['States/UTs'] != 'Total']['States/UTs']
n
np.array(df2[df2['States/UTs'] == 'Total'])[0][1:-1]
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

fig = plt.figure(figsize = (30, 15))
ax1 = fig.add_subplot(111)

for i in n:
    ax1.scatter(np.array(m)[1:-1], np.array(df2[df2['States/UTs'] == i])[0][1:-1], label=i);
    
ax1.plot(np.array(m)[1:-1], np.array(df2[df2['States/UTs'] == 'Total'])[0][1:-1]/ n.shape[0] )
plt.legend(loc='upper right')
plt.show()
```
